```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
# from matplotlib import rc
plt.rcParams.update({
    "pgf.texsystem": "pdflatex",
    "pgf.preamble": [
         r"\usepackage[utf8x]{inputenc}",
         r"\usepackage[T1]{fontenc}",
         r"\usepackage{cmbright}",
         ]
})

```

```python
file='Jac_data.xlsx'
base_loc='../generated_data/BUOY_16/exp16420'
hi=0.7;Ca=0.00012
loc=base_loc+'/'+str(hi)+'/'+str(Ca)+'/'

D=pd.read_excel(loc+file)
limit=None
Cwvec=np.array(D['Cwvec'])[:limit]
Jac1=np.array(D['Jac1'])[:limit];Jac2=np.array(D['Jac2'])[:limit]
Jac3=np.array(D['Jac3'])[:limit];Jac4=np.array(D['Jac4'])[:limit]
Jac5=np.array(D['Jac5'])[:limit]
```

```python
Jc=np.zeros((len(Cwvec),5))
for i in range(len(Cwvec)):
    Jc[i,:]=[Jac1[i],Jac2[i],Jac3[i],Jac4[i],Jac5[i]]
```

```python
name='Cost_Function'
fig=plt.figure(figsize=(12,7),frameon=True)
plt.plot(Cwvec,Jac1,'r--',Cwvec,Jac2,'--')
plt.plot(Cwvec,Jac3,'--',Cwvec,Jac4,'--')
plt.plot(Cwvec,Jac5,'--')
plt.xlabel('C_w')
plt.ylabel('Cost Function')
plt.savefig(loc+name+'.jpg',format='jpg')
plt.show()
plt.close(fig)
```

```python
name='Cost_Function_box_plot'
fig=plt.figure(figsize=(12,7),frameon=True)
plt.boxplot(Jc.T)
plt.xlabel('C_w')
plt.ylabel('Cost Function')
plt.savefig(loc+name+'.jpg',format='jpg')
plt.show()
plt.close(fig)
```
